Again,not much correlation  from these histograms. We will use scatter matrix from pandas to see if we can draw any insights from it.

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(recent_grads[['Sample_size','Median']],figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']],figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Men','ShareWomen','Median']],figsize=(10,10))
```

We are not seeing much correlation between these plots.There is a weak  negative correlation between Median and  ShareWomen Majors with less Women  tend to have higher earnings.It could be due to fact that high paying majors like Enginnering tend to have less Women.

The first ten rows are Engineering Majors and last ten rows are non-engineering majors let us plot a graph between ShareWomen and Major to see if our hypothesis is correct.

```python
recent_grads[:10].plot(kind='bar',x='Major',y='ShareWomen',colormap='winter')
```

```python
recent_grads[163:].plot(kind='bar',x='Major',y='ShareWomen',colormap='winter')
```

```python
recent_grads[:10].plot(kind='bar',x='Major',y='Median',colormap='winter')
recent_grads[163:].plot(kind='bar',x='Major',y='Median',colormap='winter')

```

Our hypothesis appears to be correct Majors with less Women such as Engineering tend to earn more.

```python
recent_grads[:10].plot(kind='bar',x='Major',y='Unemployment_rate',colormap='winter')
recent_grads[163:].plot(kind='bar',x='Major',y='Unemployment_rate',colormap='winter')
```

We plotted a graph to compare Unemployment_rate from the 10 highest paying majors and  from  the 10  lowest paying majors non-engineering majors tend have more unemployment_rate compared to Engineering majors.
